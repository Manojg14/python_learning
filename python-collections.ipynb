## List []

```python
a = [1,2,3,4,5]
print(a)
```

### Append

```python
a.append(8)
a
```

```python
a.append("Manoj")
a
```

### Insert

```python
a.insert(0,11) # here 0 is index refres and 11 is value
a
```

### Replace

```python
a[1] = 'Apple'
a
```

### Pop

```python
a.pop() # Remove the last value 
a
```

```python
a.pop(6) # here 6 refers to the index value 
a
```

### Extend

```python
b = [12,3,4,5] # combine two list 
a.extend(b)
print(a)
```

### Tuple ()

```python
a = (1,2,3,4,5,6)
a
```

```python
# Convert Tuples into List
b = list(a)
print(b)
```

## Dictionary {}

```python
a = {
    "name" : "Manoj",
    "age" : 21,
    "Dept" : "ECE",
    "Students" : ['ramu','somu','ramesh']
}
print(a)
```

```python
a.keys()
```

```python
a.values()
```

```python
a['age'] = 22
print(a)
```

```python
a['CGPA'] = 8.2
print(a)
```

### Update({})

```python
a.update({'age':23})
a
```

### Pop()

```python
a.pop("Dept")
print(a)
```

```python
del a['CGPA']
print(a)
```

### Empty Dic

```python
a.clear()
print(a)
```

### Delete Dic

```python
del a
```
