```python
def add():
    a = int(input("enter a value:"))
    b = int(input("enter b value:"))
    print(f" Addition is {a+b}")
    
def sub():
    a = int(input("enter a value:"))
    b = int(input("enter b value:"))
    print(f" Subtraction is {a-b}")
    
def mul():
    a = int(input("enter a value:"))
    b = int(input("enter b value:"))
    print(f" Multiplication is {a*b}")

def div():
    a = int(input("enter a value:"))
    b = int(input("enter b value:"))
    print(f" Division is {a/b}")
```

```python
def evenorodd(a):
    if a % 2 == 0:
        print(f"{a}, is Even Number")
    else:
        print(f"{a}, is Odd Number")

a = int(input("Enter a Value"))

evenorodd(a)
```

```python
def findpassorfail(num):
    if num > 35:
        print(f" {num} is Pass ")
    else:
        print(f" {num} is Fail")
        
number = int(input("enter Mark:"))

findpassorfail(number)
```

```python
def findrange(a,b):
    for i in range(a,b):
        print(i,end=",")

a = int(input("enter a:"))
b = int(input("enter b:"))

findrange(a,b)
```

### Function used Another Function

```python
a = int(input("enter a:"))
b = int(input("enter b:"))
c = int(input("enter c:"))

def sumofaandb(a,b):
    return a+b

# Function used with another function

def multiplywithc(c): 
    value = sumofaandb(a,b)*c
    return value

print(multiplywithc(c))
```
